```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data1=pd.read_excel("E:/DSBDA/gold2021.xlsx")
data2=pd.read_excel("E:/DSBDA/gold2020.xlsx")
data3=pd.read_excel("E:/DSBDA/gold2019.xlsx")
data4=pd.read_excel("E:/DSBDA/gold2018.xlsx")
```

```python
data=pd.concat([data1,data2,data3,data4])
data=data.sort_values(by='DATE', ascending=True)
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
data=data.set_index('DATE')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib

df1=data.sort_values(by='DATE', ascending=True)
matplotlib.rc('figure', figsize=(7, 7))
decompose = seasonal_decompose(df1.Price, period=12, model='additive')
decompose.plot()
plt.show() 
```

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
X = data.values
X = X.astype('float32')
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(X, order=(0, 1, 0))
model_fit = model.fit()
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_predict

forecast = model_fit.forecast(steps=10)[0]
plot_predict(model_fit, start=1, end=800)
plt.show()
```
